```python
def convert(obj):
    l = []
    for i in obj:
        l.append(i['name'])
    return l

```

```python
mov['genres'].apply(convert)
```

For the above issue we need to convert string to list 
so we need to use below function

```python
import ast
#ast.literal_eval()
```

```python
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l

```

```python
mov['genres'].apply(convert) # here i am getting all the genres of movies 
```

```python
mov['genres'] = mov['genres'].apply(convert)
```

now same will be applied to other columns


```python
mov['keywords'] = mov['keywords'].apply(convert)
```

```python
# now we will only take 3 dictionary from cast as start 3 Characters are the leads

def convert3(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l

mov['cast'] = mov['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
mov['crew'] = mov['crew'].apply(fetch_director)
```

```python
mov.head()
```

Since overview is string need to convert to list so that we can merge all

```python
mov['overview'].head()
```

```python
mov['overview'] = mov['overview'].apply(lambda x:x.split())
```

```python
mov['overview'].head()
```

Now we need to remove any spaces in between names in each column. Why?
Because if some search e.g. for Sam & he didn't mention the complete name so model will get confused

```python
mov['genres'] = mov['genres'].apply(lambda x:[i.replace(" ","") for i in x])
mov['keywords'] = mov['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
mov['cast'] = mov['cast'].apply(lambda x:[i.replace(" ","") for i in x])
mov['crew'] = mov['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
mov.head()
```

```python
# Now we will concatenate below columns


mov['tags'] = mov['overview'] + mov['genres']+ mov['cast']+ mov['crew']+ mov['keywords']
```

#Vectorization
We need to convert text into vectors for which we are going to use technique Bags of words.
Hence we will combine all tags into one and then find 5000 most common words(it depends on us to choose).More the words more the dimentionality

Then we will take each tag & count the frequency of occuring of each word in each tag by which we will get 5000*5000 dataframe in which each row is the vector for movie.

These vectors will be plotted in 2d....5000D space.

Stop words(use in sentence formation) are also removed while vectorization eg to, all, the, if.....

The above complete process will be done scikit learn class - countvectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_data['tags']).toarray()
#converting to array because countvec..return sparse matrix
```

```python
vectors.shape
```

```python
vectors
```

```python
cv.get_feature_names()
```

above data has one issue withn it i.e. words are repeating like love,loves,loving,loved...
Hence we will apply stemming which will convert above words....into one word.

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps =PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():# converting to list
        y.append(ps.stem(i))
    
    str = " ".join(y)
    
    return str
        
```

```python
new_data['tags'][0]
```

```python
new_data['tags'] = new_data['tags'].apply(stem)
```

now we have 4806 movies vector. These vectors will be plotted in n-d space and need to get the distance between each vector.

In order to this we won't use euclidian distance(due to high dimention) therefore we are going to use cosine distance ( angle btw vectors )

```python
from sklearn.metrics.pairwise import cosine_similarity# btw 0 & 1
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
# below is the array of array which gives / provides distance between movie from each movies
similarity
```

```python
def recommend(movie):
    x = new_data[new_data['title'] == movie].index[0] # fetching index of movie
    distance = similarity[x]  # Here we got the ditance of the movie from each movies as array
    mov_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in mov_list:
        print (new_data.iloc[i[0]].title)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6] # below is tuple
```

```python
recommend('Spectre')
```
